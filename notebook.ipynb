```python
def get_player_rating(username):
    data = get_player_stats(username).json['stats']
    print(data)
    categories = ['chess_blitz', 'chess_rapid', 'chess_bullet']

    for category in categories:
        print('Category:', category)
        print(f'Current: {data[category]["last"]["rating"]}')

    printer.pprint(data)
```

```python
import pandas as pd

# Define an empty list to store the parsed games
parsed_games = []

# Open the PGN file and read line by line
with open('example.pgn') as f:
    game = {}
    for line in f:
        line = line.strip()
        if line.startswith('['):
            # Parse the header tags and store in the game dictionary
            tag, value = line[1:-1].split(' ')
            game[tag] = value
        elif line == '':
            # The end of game is reached, add the game dictionary to the list of games
            parsed_games.append(game)
            game = {}
        else:
            # Parse the moves and store in the game dictionary
            game['moves'] = line

# Create a pandas dataframe from the parsed games
df = pd.DataFrame(parsed_games)

# Print the dataframe
print(df.head())
```
